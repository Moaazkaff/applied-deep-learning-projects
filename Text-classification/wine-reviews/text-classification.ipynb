```python
# grabe the file and load it as a table.
# read_csv convert the comman seperated values into dataframes.
df_imdb = pd.read_csv('IMDB Dataset.csv')
```

```python
# head func shows the first 5 rows.
df_imdb.head()
```

```python
# Why List Comprehension Works and the For loop fails , 
# Because the list comprehension creates a new list and assigns it to the 'sentiment' column, while the for loop tries to modify the existing 'sentiment' column in place, which is not allowed.
# In list comprehension, pandas isn't involved in the looping process at all.
# When you use a for loop, you are trying to modify the DataFrame piece-by-piece, which triggers several mistakes. 
df_imdb['sentiment'] = [1 if x == 'positive' else 0 for x in df_imdb['sentiment']]
df_imdb.head(5)
```

```python
# Values count method job is to look at a column of data, find all the unique items, and count exactly how many times each item appears.
# make sure the words actually turned into numbers.
print(df_imdb['sentiment'].value_counts())
```
